```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
L = 1
g = 9.8
theta0 = 0.1
omega0 = 1.0
b_m = 0.2
t = np.linspace(0,20,1000)
def linear_pendulum(y,t):
theta, omega = y
dtheta_dt = omega
domega_dt = -(g/L)*theta
return [dtheta_dt,domega_dt]
y0=[theta0,omega0]
sol1=odeint(linear_pendulum,y0,t)
theta_linear=sol1[:,0]
omega_linear=sol1[:,1]
plt.plot(t,theta_linear)
plt.title("Small Angle Pendulum")
plt.xlabel("Time")
plt.ylabel("Angular Displacement")
plt.savefig("small_angle_pendulum.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def nonlinear_pendulum(y,t):
theta,omega=y
dtheta_dt=omega
domega_dt=-(g/L)*np.sin(theta)
return [dtheta_dt,domega_dt]
sol2=odeint(nonlinear_pendulum,y0,t)
theta_nonlinear=sol2[:,0]
omega_nonlinear=sol2[:,1]
plt.plot(t,omega_nonlinear)
plt.title("Angular Velocity (Nonlinear)")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.savefig("nonlinear_pendulum.png")
plt.show()
plt.plot(t,theta_linear,label="Linear")
plt.plot(t,theta_nonlinear,label="Nonlinear")
plt.legend(loc='upper right')
plt.title("Linear vs Nonlinear Comparison")
plt.xlabel("Time")
plt.ylabel("Angular Displacement")
plt.savefig("linear_vs_nonlinear.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def damped_pendulum(y,t):
theta,omega=y
dtheta_dt=omega
domega_dt=-(b_m*omega)-(g/L)*np.sin(theta)
return [dtheta_dt,domega_dt]
sol3=odeint(damped_pendulum,y0,t)
theta_damped=sol3[:,0]
omega_damped=sol3[:,1]
plt.plot(t,theta_damped)
plt.title("Damped Pendulum")
plt.xlabel("Time")
plt.ylabel("Angular Displacement")
plt.savefig("damped_pendulum.png")
plt.show()
plt.plot(t,omega_damped)
plt.title("Angular Velocity (Damped)")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.savefig("damped_pendulum.png")
plt.show()
```
