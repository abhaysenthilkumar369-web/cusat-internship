## 12. Functions in Python

Below is an example of a function that computes the value of the expression  
**y = x² + y²**, given values for `x` and `y`.

Try below code

```
# This function computes y = x^2 + y^2
def compute_expression(x, y):
    # x**2 is x squared, y**2 is y squared
    result = x**2 + y**2
    return result

# Example: compute the value when x = 3 and y = 4
output = compute_expression(3, 4)

# Print the result
print("Result:", output)
```

- Defines a function `compute_expression` that takes two inputs, `x` and `y`
- Computes the expression \( x^2 + y^2 \)
- Returns the result
- Calls the function with `x = 3` and `y = 4`
- Prints the result


`````python
def AV_rectangle(a, b, c):
  area = a * b #local variable
  volume = a* b* c
  return area, volume

area, volume  = AV_rectangle(10, 5, 2)
print(area, volume)

`````
