# Problem 1: Numerical Riemann Integral



Compute the integral of the function f(x) = sin(x) over the interval [0, π] using the Left Riemann Sum and the Midpoint Rule. Compare the results with the analytical solution.

Computer $I = \int_0^{\pi} \sin(x) dx $.

###Midpoint formula

$
    \int_a^b f(x)\,dx \approx \sum_{i=0}^{n-1} f\!\left(x_i + \frac{\Delta x}{2}\right)\, \Delta x
$




[Figure for midpoint formula
](https://www.allmath.com/storage/2023/Dec/midpointrule_74.png)




### Try a code somewhat like the below one
```
import numpy as np
# Function and interval

a, b = 0, np.pi
n    = 10  # number of sub-intervals
dx   = (b - a)/n


# Midpoint Rule
integral = 0.0
x_val    = a + dx/2
for i in range(n):
    integral += np.sin(x_val) * dx
    x_val         = x_val + dx


# Analytical solution
integral_exact = 2.0

print(f"Midpoint Rule (for loop):   {integral:.6f}")
print(f"Analytical Integral:        {integral_exact:.6f}")

```

# Solve the Harmonic Oscillator Problem

Consider the equation for a harmonic oscillator:

$
m\frac{d^2 x}{dt^2} = -k x.
$

This can be written as

$
\dot{x} = v,\\[4pt]
\dot{v} = -\frac{k}{m}x.
$

###Descrete equations
With $k/m=1$, we have

$
x_{n+1} = x_n + v_n\,\Delta t,\\[4pt]
v_{n+1} = v_n - x_n\,\Delta t
$


### Try below code

```
import numpy as np
import matplotlib.pyplot as plt

# Parameters
dt = 0.01                  # time step
t_max = 8 * np.pi          # end time
t = np.arange(0, t_max + dt, dt)  # time array
n = len(t)

# Arrays to store x(t) and v(t)
x = np.zeros(n)
v = np.zeros(n)

# Initial conditions
x[0] = 0.0
v[0] = 1.0

# Forward Euler integration using for loop
for i in range(n-1):
    x[i+1] = x[i] + v[i] * dt
    v[i+1] = v[i] - x[i] * dt

# Analytical solutions
x_exact = np.sin(t)
v_exact = np.cos(t)

# Plot x(t) comparison
plt.figure(figsize=(8,4))
plt.plot(t, x, label='Numerical x(t)', lw=2)
plt.plot(t, x_exact, '--', label='Analytical x(t) = sin(t)')
plt.xlabel('Time t')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
plt.show()

```

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
dt = 0.01                  # time step
t_max = 8 * np.pi          # end time
t = np.arange(0, t_max + dt, dt)  # time array
n = len(t)

# Arrays to store x(t) and v(t)
x = np.zeros(n)
v = np.zeros(n)

# Initial conditions
x[0] = 0.0
v[0] = 1.0

# Forward Euler integration using for loop
for i in range(n-1):
    x[i+1] = x[i] + v[i] * dt
    v[i+1] = v[i] - x[i] * dt

# Analytical solutions
x_exact = np.sin(t)
v_exact = np.cos(t)

# Plot x(t) comparison
plt.figure(figsize=(8,4))
plt.plot(t, x, label='Numerical x(t)', lw=2)
plt.plot(t, x_exact, '--', label='Analytical x(t) = sin(t)')
plt.xlabel('Time t')
plt.ylabel('x(t)')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
````
